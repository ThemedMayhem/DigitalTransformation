```python
data = pd.read_csv('TNVED.csv', sep='\t')
```

```python
for i in range(len(data.KOD)):
    if data.KOD[i].find('X') != -1:
        data = data.drop([i], axis=0)
data = data.reset_index(drop=True)
```

```python
data
```

```python
data['len_kod'] = data.KOD.apply(lambda x: len(x))
data
```

```python
data['len_kod'].unique()
```

```python
data_2kod = data[data.len_kod == 2]
data_2kod = data_2kod.drop(['len_kod'], axis=1)
data_2kod
```

```python
data_4kod = data[data.len_kod == 4]
data_4kod = data_4kod.drop(['len_kod'], axis=1)
data_4kod
```

```python
data_6kod = data[data.len_kod == 6]
data_6kod = data_6kod.drop(['len_kod'], axis=1)
data_6kod
```

```python
data_8kod = data[data.len_kod == 8]
data_8kod = data_8kod.drop(['len_kod'], axis=1)
data_8kod
```

```python
data_10kod = data[data.len_kod == 10]
data_10kod = data_10kod.drop(['len_kod'], axis=1)
data_10kod
```

```python
data_2kod.to_csv('data_2kod.csv')
data_4kod.to_csv('data_4kod.csv')
data_6kod.to_csv('data_6kod.csv')
data_8kod.to_csv('data_8kod.csv')
data_10kod.to_csv('data_10kod.csv')
```
